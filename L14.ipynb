```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

#step 1
titanic = pd.read_csv('titanic_data.csv')

#step 2
print(titanic.describe())

plt.hist(titanic['Pclass'], bins=3)
plt.title('Passenger class of people')
plt.show()
plt.hist(titanic['Age'], bins=10)
plt.title('Age of people')
plt.show()
plt.plot(titanic['Age'], titanic['Pclass'], '.')
plt.title('Passenger class vs age')
plt.show()
plt.plot(titanic['Pclass'], titanic['Survived'], '.')
plt.title('Passenger class vs survival')
plt.show()
plt.hist(titanic[titanic['Survived'] == 1]['Pclass'], bins=3)
plt.title('Passenger class of people who survived')
plt.show()
plt.hist(titanic[titanic['Survived'] == 1]['Age'])
plt.title('Age of people who survived')
plt.show()
plt.hist(titanic[titanic['Survived'] == 0]['Age'])
plt.title('Age of people who no survived')
plt.show()
plt.hist(titanic[titanic['Survived'] == 0]['Pclass'], bins=3)
plt.title('Passenger class of people who no survived')
plt.show()
plt.scatter(titanic['Pclass'], titanic['Fare'])
plt.title('Passenger class versus the fare they paid')
plt.show()
plt.scatter(titanic['Survived'], titanic['Siblings/Spouses Aboard'])
plt.title('Survival vs how many siblings/spouses aboard')
plt.show()
plt.scatter(titanic['Survived'], titanic['Parents/Children Aboard'])
plt.title('Survival vs how many parents/children aboard')
plt.show()
plt.hist(titanic['Parents/Children Aboard'])
plt.title('How many parents/children aboard per person')
plt.show()
plt.hist(titanic['Siblings/Spouses Aboard'])
plt.title('How many siblings/spouses aboard per person')
plt.show()
plt.hist(titanic[titanic['Survived'] == 1]['Parents/Children Aboard'])
plt.title('How many parents/children aboard per person who survived')
plt.show()
plt.hist(titanic[titanic['Survived'] == 1]['Siblings/Spouses Aboard'])
plt.title('How many siblings/spouses aboard per person who survived')
plt.show()
```

4. The 'Survived' column appears to represent our target.

5. Our hypothesis is that survivability is a function of class and age. The better your class, the higher your survivability. If your age falls within 20's or 30's, you are more likely to survive.
